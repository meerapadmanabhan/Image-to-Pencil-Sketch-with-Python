### IMPORT LIBRARIES

```python
import matplotlib.pyplot as plt
import cv2
```

### IMPORT IMAGE

```python
image= cv2.imread("lab.jpg")
```

### DISPLAY IMAGE

```python
plt.imshow(image)
plt.show()
```

### RGB FORM

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.show()
```

### ORIGINAL TO GREYSCALE

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image
```

```python
plt.imshow(gray_image)
plt.show()
```

### INVERTED IMAGE

```python
inverted_image = 255 - gray_image
cv2.imshow("Inverted", inverted_image)
cv2.waitKey()
```

```python
plt.imshow(inverted_image)
plt.show()
```

### BLUR IMAGE

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
plt.imshow(blurred)
plt.show()
```

### PENCIL SKETCH

```python
pencil_sketch = cv2.divide(gray_image, 255 - blurred, scale=256)
plt.imshow(pencil_sketch, cmap='gray')
plt.show()
```

```python
plt.imshow(pencil_sketch)
plt.show()
```
